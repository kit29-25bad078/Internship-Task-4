```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    "Order_ID": [1001, 1002, 1003, 1004, 1005, 1006, 1007, 1008],
    "Product": ["Mobile", "Laptop", "Headphones", "Smartwatch",
                "Tablet", "Mobile", "Laptop", "Headphones"],
    "Quantity": [2, 1, 3, 2, 1, 4, 2, 5],
    "Price": [15000, 55000, 2000, 8000, 25000, 15000, 55000, 2000]
}

df = pd.DataFrame(data)


df["Revenue"] = df["Quantity"] * df["Price"]

print("========== E-COMMERCE ANALYSIS REPORT ==========\n")
print(df)

print("\n========== KPI REPORT ==========")
print("Total Orders:", len(df))
print("Total Revenue:", df["Revenue"].sum())
print("Average Revenue:", round(df["Revenue"].mean(), 2))
print("Highest Revenue:", df["Revenue"].max())
print("Lowest Revenue:", df["Revenue"].min())


product_revenue = df.groupby("Product")["Revenue"].sum()
print("\nProduct Wise Revenue")
print(product_revenue)


best_product = df.groupby("Product")["Quantity"].sum().idxmax()
print("\nBest Selling Product:", best_product)


print("\n========== SUMMARY ==========")
print(df.describe())


plt.figure(figsize=(8,5))
plt.bar(df["Product"], df["Revenue"])
plt.title("Product Revenue Analysis")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()


plt.figure(figsize=(8,5))
plt.plot(df["Product"], df["Quantity"], marker='o')
plt.title("Product Sales Trend")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.grid(True)
plt.show()

print("\nE-Commerce Analysis Completed Successfully!")
```
